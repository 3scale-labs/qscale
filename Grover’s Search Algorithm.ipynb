```python
# This is a quantum representation in a circuit for the "Find the Queen" problem
# In this case, the queen is always placed at the 4th of 4 position (cards)
# The quantum computing finds the queen always in one single shot,
# whereas a classical computer would require 2.25 attempts in average

from qiskit import *
%matplotlib inline
```

```python
circuit = QuantumCircuit(2, 2)
circuit.h(0)
circuit.h(1)

circuit.h(1)
circuit.cx(0,1)
circuit.h(1)
circuit.barrier()

circuit.h(0)
circuit.h(1)
circuit.x(0)
circuit.x(1)

circuit.h(1)
circuit.cx(0,1)
circuit.h(1)
circuit.barrier()

circuit.x(0)
circuit.x(1)
circuit.h(0)
circuit.h(1)

circuit.measure([0,1], [0,1])
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1).result()
counts = result.get_counts()
print(counts)
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
